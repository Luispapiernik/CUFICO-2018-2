```
import ROOT
```

```
from array import array
```

```
MyTree = ROOT.TTree( 'tree1', 'tree para dos variables con distribucion gausiana y una variable con distribucion constante' )
```

```
x = array('f', [0])
y = array('f', [0])
z = array('f', [0])
```

```
MyTree.Branch('x', x, 'x/F')
MyTree.Branch('y', y, 'y/F')
MyTree.Branch('z', z, 'z/F')
```

```
import numpy as np
```

```
for i in range(1000):
  # normal con media 5 y desviazion 2
  tempx = np.random.normal(5, 2)
  tempy = np.random.normal(5, 2)
  # todos los valores deben estar entre 0 y 10
  if -1 < tempx < 11 and -1 < tempy < 11:
    x[0] = tempx
    y[0] = tempy
    z[0] = np.random.rand() * 10
    MyTree.Fill()
```

```
c1 = ROOT.TCanvas( 'c1', 'Graficas para combinaciones de "x", "y" y "z"', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x * y >> h1(50,0,100)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x * z >> h2(50,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x * y * z>>h3(50, 0, 1000)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
# no tiene sentido comparar igualdad en reales, mayor y menor que tiene mas sentido
MyTree.Draw("x >> h4(10,0,10)","z - 5 < 0.01")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
# no tiene sentido comparar igualdad en reales, mayor y menor que tiene mas sentido
MyTree.Draw("x:z >> h5(10,0,10,10,0,10)","y - 5 < 0.01", "lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
# no tiene sentido comparar igualdad en reales, mayor y menor que tiene mas sentido
MyTree.Draw("x:y:z >> h6(20,0,10,20,0,10,20,0,10)", "", "lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
